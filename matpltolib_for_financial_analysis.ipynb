```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import yfinance as yf
```

```python
tickers = ["GOOG", "AMZN", "GLD"]
df = yf.download(tickers=tickers, start = "2019-08-01")["Close"]
df.head()
```

```python
df.head()
```

```python
returns = np.log(df).diff()
returns.head()
returns.dropna(inplace=True)
```
